## Описание данных

Данные описывают информацию о бронировании билетов на сервисе Яндекс.Афиша за период с 1 июня по 31 октября 2024 года. Анализ проводится с целью выявления сезонных изменений в поведении пользователей, популярности событий и особенностях использования мобильных и стационарных устройств.

Данные представлены в трёх файлах:

- **final_tickets_orders_df.csv** (информация о заказах):
  Этот файл содержит данные о каждом заказе, оформленном пользователями сервиса.
  - `order_id` — уникальный идентификатор заказа;
  - `user_id` — уникальный идентификатор пользователя;
  - `created_dt_msk` — дата создания заказа (по московскому времени);
  - `created_ts_msk` — полная дата и время создания заказа;
  - `event_id` — идентификатор мероприятия, на которое были приобретены билеты;
  - `cinema_circuit` — сеть кинотеатров, если применимо (иначе значение "нет");
  - `age_limit` — возрастное ограничение мероприятия;
  - `currency_code` — валюта оплаты (rub — российский рубль, kzt — казахстанский тенге);
  - `device_type_canonical` — тип устройства, с которого был сделан заказ (mobile или desktop);
  - `revenue` — выручка по заказу;
  - `service_name` — название билетного оператора;
  - `tickets_count` — количество купленных билетов;
  - `total` — общая сумма заказа;
  - `days_since_prev` — количество дней с момента предыдущего заказа того же пользователя (может содержать пропуски для первого заказа).

- **final_tickets_events_df.csv** (информация о мероприятиях):
  Файл содержит данные о мероприятиях, доступных для бронирования. В этом датасете исключены фильмы, так как их доля в данных невелика.
  - `event_id` — уникальный идентификатор мероприятия;
  - `event_name` — название мероприятия;
  - `event_type_description` — описание типа мероприятия;
  - `event_type_main` — основной тип мероприятия (например, концерт, театр, спорт, выставки);
  - `organizers` — организаторы мероприятия;
  - `region_name` — регион проведения мероприятия;
  - `city_name` — город проведения мероприятия;
  - `city_id` — идентификатор города;
  - `venue_id` — уникальный идентификатор площадки;
  - `venue_name` — название площадки;
  - `venue_address` — адрес места проведения мероприятия.

- **final_tickets_tenge_df.csv** (курс тенге к рублю):
  Этот датасет используется для конвертации выручки из казахстанских тенге в российские рубли.
  - `data` — дата;
  - `nominal` — номинал валюты (100 тенге);
  - `curs` — курс тенге к рублю на указанную дату;
  - `cdx` — код валюты (kzt).

```python
# Преобразуем created_dt_msk и created_ts_msk в datetime
orders['created_dt_msk'] = pd.to_datetime(orders['created_dt_msk'])
orders['created_ts_msk'] = pd.to_datetime(orders['created_ts_msk'])
```

```python
# Проверка успешности преобразования в datetime
print("Типы данных после преобразования:")
print(orders[['created_dt_msk', 'created_ts_msk']].dtypes)
```

```python
# Создание новых столбцов:
# a) revenue_rub
# Функция для конвертации валюты
def convert_to_rub(row):
    if row['currency_code'] == 'rub':
        return row['revenue']
    elif row['currency_code'] == 'kzt':
        # Находим курс на дату заказа
        rate_date = row['created_dt_msk'].date()
        rate = tenge_rates[tenge_rates['data'] == str(rate_date)]['curs'].values
        if len(rate) > 0:
            return row['revenue'] * rate[0] / 100  # Конвертируем тенге в рубли
        else:
            return np.nan  # Если курс не найден
    else:
        return np.nan
```

```python
# Применяем функцию
orders['revenue_rub'] = orders.apply(convert_to_rub, axis=1)
```

```python
# b) one_ticket_revenue_rub
orders['one_ticket_revenue_rub'] = orders['revenue_rub'] / orders['tickets_count']
```

```python
# c) month
orders['month'] = orders['created_dt_msk'].dt.month
```

```python
# d) season
def get_season(month):
    if month in [12, 1, 2]:
        return 'зима'
    elif month in [3, 4, 5]:
        return 'весна'
    elif month in [6, 7, 8]:
        return 'лето'
    else:
        return 'осень'

orders['season'] = orders['month'].apply(get_season)
```

```python
# Изменение средней стоимости билета по типам мероприятий
summer_revenue = summer.groupby('event_type_main')['one_ticket_revenue_rub'].mean().sort_values(ascending=False)
autumn_revenue = autumn.groupby('event_type_main')['one_ticket_revenue_rub'].mean().sort_values(ascending=False)

comparison = pd.DataFrame({
    'summer': summer_revenue,
    'autumn': autumn_revenue
}).dropna()

# Относительное изменение
comparison['change'] = (comparison['autumn'] - comparison['summer']) / comparison['summer'] * 100

print("Сравнение средней стоимости билета летом и осенью:")
display(comparison.sort_values(by='change', ascending=False))
```

```python
# Cравнение количества заказов для разных типов мероприятий
# Группируем данные по сезону и типу мероприятия
season_event_counts = merged_df.groupby(['season', 'event_type_main']).size().reset_index(name='order_count')

# Рассчитываем общее количество заказов по сезонам
total_orders_by_season = season_event_counts.groupby('season')['order_count'].sum()

# Вычисляем доли для каждого типа мероприятия в каждом сезоне
season_event_counts['share'] = season_event_counts.apply(
    lambda row: row['order_count'] / total_orders_by_season[row['season']], axis=1
)

# Выводим результаты
print("Cравнение количества заказов для разных типов мероприятий:")
display(season_event_counts)
```

```python
# Создаём сводную таблицу с долями
pivot_table = season_event_counts.pivot(index='event_type_main', columns='season', values='share').fillna(0)

# Визуализация
ax = pivot_table.plot(kind='bar', figsize=(12, 6))
plt.title('Доля заказов по типам мероприятий летом и осенью', pad=20)
plt.xlabel('Тип мероприятия')
plt.ylabel('Доля заказов')
plt.legend(title='Сезон')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)  # Горизонтальные подписи

# Добавляем значения на столбцы (все подписи сверху)
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Показываем только положительные значения
        ax.annotate(
            f'{height:.1%}',  # Форматируем как проценты с 1 знаком после запятой
            (p.get_x() + p.get_width() / 2., height),
            ha='center',
            va='bottom',  # Все подписи сверху столбцов
            fontsize=9,
            color='black',
            xytext=(0, 3),  # Смещение подписи вверх
            textcoords='offset points'
        )

plt.grid(False)
plt.tight_layout()
plt.show()
```

Осенью растет спрос на театр и спорт, тогда как летом выше доля концертов и разовых мероприятий ("другое"). Сезонность четко прослеживается.

### 3.2. Осенняя активность пользователей

```python
# Фильтруем данные только за осень (сентябрь–октябрь)
autumn_data = merged_df[merged_df['month'].isin([9, 10])].copy()
```

```python
# Добавляем день недели (0 — понедельник, 6 — воскресенье)
autumn_data['day_of_week'] = autumn_data['created_dt_msk'].dt.dayofweek
```

```python
# Агрегируем данные по дням
daily_stats = autumn_data.groupby('created_dt_msk').agg(
    orders=('order_id', 'count'),
    users=('user_id', 'nunique'),
    avg_ticket_price=('one_ticket_revenue_rub', 'mean')
).reset_index()
```

```python
# Рассчитываем среднее количество заказов на пользователя
daily_stats['orders_per_user'] = daily_stats['orders'] / daily_stats['users']
```

```python
# Графики динамики
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Функция для добавления пиковых значений
def add_peak_values(ax, data, x_col, y_col, color='black'):
    peaks = data.nlargest(3, y_col)
    for x, y in zip(peaks[x_col], peaks[y_col]):
        ax.text(x, y, f'{int(y) if y_col != "orders_per_user" else f"{y:.2f}"}', 
                ha='center', va='bottom', fontsize=9,
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=1),
                color=color)

# 1. Число заказов
ax1 = sns.lineplot(data=daily_stats, x='created_dt_msk', y='orders', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Число заказов по дням', pad=15)
axes[0, 0].set_xlabel('Дата', labelpad=10)
axes[0, 0].set_ylabel('Количество заказов', labelpad=10)
axes[0, 0].grid(False)
add_peak_values(axes[0, 0], daily_stats, 'created_dt_msk', 'orders', 'skyblue')

# 2. DAU
ax2 = sns.lineplot(data=daily_stats, x='created_dt_msk', y='users', ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Количество уникальных пользователей по дням', pad=15)
axes[0, 1].set_xlabel('Дата', labelpad=10)
axes[0, 1].set_ylabel('Количество пользователей', labelpad=10)
axes[0, 1].grid(False)
add_peak_values(axes[0, 1], daily_stats, 'created_dt_msk', 'users', 'salmon')

# 3. Заказы на пользователя
ax3 = sns.lineplot(data=daily_stats, x='created_dt_msk', y='orders_per_user', ax=axes[1, 0], color='teal')
axes[1, 0].set_title('Среднее число заказов на пользователя', pad=15)
axes[1, 0].set_xlabel('Дата', labelpad=10)
axes[1, 0].set_ylabel('Заказов на пользователя', labelpad=10)
axes[1, 0].grid(False)
add_peak_values(axes[1, 0], daily_stats, 'created_dt_msk', 'orders_per_user', 'teal')

# 4. Средняя стоимость билета
ax4 = sns.lineplot(data=daily_stats, x='created_dt_msk', y='avg_ticket_price', ax=axes[1, 1], color='goldenrod')
axes[1, 1].set_title('Средняя стоимость билета', pad=15)
axes[1, 1].set_xlabel('Дата', labelpad=10)
axes[1, 1].set_ylabel('Средняя стоимость (руб)', labelpad=10)
axes[1, 1].grid(False)
add_peak_values(axes[1, 1], daily_stats, 'created_dt_msk', 'avg_ticket_price', 'goldenrod')

# Поворачиваем подписи дат на 45 градусов для всех графиков
for ax in axes.flat:
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
        tick.set_ha('right')

plt.tight_layout()
plt.show()
```

Краткие выводы по графикам:
- Количество заказов демонстрирует недельную цикличность, с пиками в выходные дни (суббота и воскресенье). Общая тенденция к росту числа заказов от сентября до октября.
- Активность пользователей также имеет недельную цикличность, с пиками в выходные дни. Наблюдается стабильный рост общего числа уникальных пользователей с сентября по октябрь.
- Среднее число заказов на одного пользователя колеблется в диапазоне от 4 до 7. В целом, среднее число заказов на пользователя остаётся стабильным, без значительных изменений.
- Средняя стоимость билета колеблется в диапазоне от 80 до 250 рублей. В целом, средняя стоимость билета сохраняет относительно стабильный уровень, хотя наблюдаются небольшие отклонения.

## 4. Статистический анализ данных и проверка гипотез

Проверим две гипотезы , связанные с пользовательским поведением на мобильных и стационарных устройствах: 
* Гипотеза 1 (H₁): Среднее количество заказов на одного пользователя мобильного приложения выше, чем у пользователей стационарных устройств. 
* Гипотеза 2 (H₁): Среднее время между заказами у пользователей мобильных устройств больше, чем у пользователей стационарных устройств.

```python
# Настройка графиков
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Подготовка данных
# Оставляем только осенние данные (сентябрь–октябрь)
autumn_data = merged_df[merged_df['month'].isin([9, 10])]
```

```python
# Среднее число заказов на пользователя по типу устройства
orders_per_user = autumn_data.groupby(['user_id', 'device_type_canonical']).size().reset_index(name='order_count')
```

```python
# Среднее время между заказами (days_since_prev)
time_between_orders = autumn_data.dropna(subset=['days_since_prev'])
```

### 4.1. Проверка первой гипотезы

```python
# Гипотеза 1: 
# H₀: Среднее число заказов на пользователя одинаково для мобильных и стационарных устройств.
# H₁: Среднее число заказов на пользователя выше у пользователей мобильных устройств.

# Разделяем данные по типу устройства
mobile_orders = orders_per_user[orders_per_user['device_type_canonical'] == 'mobile']['order_count']
desktop_orders = orders_per_user[orders_per_user['device_type_canonical'] == 'desktop']['order_count']
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(mobile_orders[mobile_orders < 10], color='skyblue', label='Mobile', bins=30, kde=True)
sns.histplot(desktop_orders[desktop_orders < 10], color='salmon', label='Desktop', bins=30, kde=True)
plt.title('Распределение числа заказов (без выбросов)')
plt.xlabel('Число заказов (<10)')
plt.ylabel('Количество пользователей')
plt.legend()
ax = plt.gca()
ax.grid(False)
plt.show()
```

Распределение числа заказов на пользователя показывает, что большинство пользователей оформляют один заказ. 

Выбор теста для проверки первой гипотезы:
 * выбран тест Манна-Уитни, так как распределение числа заказов на пользователя — ненормальное, содержит выбросы.
 * T-тест был бы менее корректным из-за чувствительности к экстремальным значениям и требованию нормальности распределения.

```python
# H₀: Среднее число заказов на пользователя одинаково для мобильных и стационарных устройств.
# H₁: Среднее число заказов на пользователя выше у мобильных устройств.

# Проверка гипотезы
stat, p_value = mannwhitneyu(mobile_orders, desktop_orders, alternative='greater')

print(f"p-value для гипотезы о большем числе заказов на мобильных: {p_value:.5f}")
if p_value < 0.05:
    print("✅ Отклоняем H₀: среднее число заказов на мобильных выше.")
else:
    print("❌ Не отклоняем H₀: нет значимой разницы.")
```

Мы отклоняем нулевую гипотезу (H₀), что среднее число заказов на пользователя одинаково для мобильных и стационарных устройств.

Среднее число заказов на одного пользователя выше у мобильных устройств .

### 4.2. Проверка второй гипотезы

```python
# Гипотеза 2:
# H₀: Среднее время между заказами одинаково для мобильных и стационарных устройств
# H₁: Среднее время между заказами больше у пользователей мобильных устройств

# Разделяем данные по типу устройства
mobile_days = time_between_orders[time_between_orders['device_type_canonical'] == 'mobile']['days_since_prev']
desktop_days = time_between_orders[time_between_orders['device_type_canonical'] == 'desktop']['days_since_prev']
```

```python
# Визуализация распределения
plt.figure(figsize=(10, 6))
sns.histplot(mobile_days[mobile_days < 30], color='skyblue', label='Mobile', bins=30, kde=True)
sns.histplot(desktop_days[desktop_days < 30], color='salmon', label='Desktop', bins=30, kde=True)

plt.title('Распределение времени между заказами')
plt.xlabel('Дни между заказами')
plt.ylabel('Количество пользователей')
plt.legend(title='Тип устройства')  
ax = plt.gca()
ax.grid(False)
plt.show()
```

Распределение времени между заказами показывает, что большинство пользователей делают заказы сразу или через короткий промежуток времени.

Для проверки второй гипотезы также использован тест Манна-Уитни, так как:
* распределение времени между заказами не является нормальным;
* присутствуют выбросы (очень большие промежутки между заказами);
* тест позволяет сравнивать медианные значения и работает с рангами, а не с сырыми данными.

```python
# H₀: Среднее время между заказами одинаково для мобильных и стационарных устройств.
# H₁: Среднее время между заказами больше у мобильных устройств.

# Проверка гипотезы
stat, p_value = mannwhitneyu(mobile_days, desktop_days, alternative='greater')

print(f"p-value для гипотезы о большем времени между заказами на мобильных: {p_value:.5f}")
if p_value < 0.05:
    print("✅ Отклоняем H₀: время между заказами на мобильных выше.")
else:
    print("❌ Не отклоняем H₀: нет значимой разницы.")
```

Мы отклоняем нулевую гипотезу (H₀), что среднее время между заказами одинаково для мобильных и стационарных устройств. 

Среднее время между заказами больше у мобильных устройств .
